### Statistical Testing
1. Since population and gdp have a natural trend of increasing over years, we need to conduct statistical analysis to determine whether legalization of weed has affected that trend.
2. Since we assume the homogeinity of variance in our pre- and post legalization population samples, we can perform a z-test.
3. Our null-hypothesis assumes that the true mean difference between the paired samples is zero which means any difference between pre- and post metrics is due to random variation.
4. Our alternative hypothesis assumes that the true mean difference between the paired samples is not equal to zero and is might be due to herb leaglization.

```python
crime = pd.read_csv('db/crime_rates.csv')
crime.columns
```

### T-test on California property crime data

```python
a = crime.loc[(crime["Year"] < 2016), ["California"]]
b = crime.loc[(crime["Year"] >= 2016), ["California"]]

a1 = a['California'].values.tolist()
b1 = b['California'].values.tolist()


# runnning a t-test
t, p = stats.ttest_ind(a1,b1)
print("t-statistics = " + str(t))
print("p = " + str(2*p))
```

```python
a = crime.loc[(crime["Year"] < 2012), ["Colorado"]]
b = crime.loc[(crime["Year"] >= 2012), ["Colorado"]]

a1 = a['Colorado'].values.tolist()
b1 = b['Colorado'].values.tolist()

# runnning a t-test
t, p = stats.ttest_ind(a1,b1)
print("t-statistics = " + str(t))
print("p = " + str(2*p))
```

```python
a = crime.loc[(crime["Year"] < 2012), ["Year", "Washington"]]
b = crime.loc[(crime["Year"] >= 2012), ["Year", "Washington"]]

a1 = a['Washington'].values.tolist()
b1 = b['Washington'].values.tolist()

# runnning a t-test
t, p = stats.ttest_ind(a1,b1)
print("t-statistics = " + str(t))
print("p = " + str(2*p))

plt.plot(a["Year"], a["Washington"], color='g')
plt.plot(b["Year"], b["Washington"], color='orange')
plt.show()
```
