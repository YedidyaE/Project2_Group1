### Statistical Testing
1. Since population and gdp have a natural trend of increasing over years, we need to conduct statistical analysis to determine whether legalization of weed has affected that trend.
2. Since we assume the homogeinity of variance in our pre- and post legalization population samples, we can perform a z-test.
3. Our null-hypothesis assumes that the true mean difference between the paired samples is zero which means any difference between pre- and post metrics is due to random variation.
4. Our alternative hypothesis assumes that the true mean difference between the paired samples is not equal to zero and is might be due to herb leaglization.

```python
#Colorado data
CO_pre_weed = pop_df.loc[(pop_df["State"] == "Colorado") & (pop_df["Year"] < 2014), ["Population", "Year"]]
CO_post_weed = pop_df.loc[(pop_df["State"] == "Colorado") & (pop_df["Year"] >= 2014), ["Population", "Year"]]

#Washington data
WA_pre_weed = pop_df.loc[(pop_df["State"] == "Washington") & (pop_df["Year"] < 2014), ["Population", "Year"]]
WA_post_weed = pop_df.loc[(pop_df["State"] == "Washington") & (pop_df["Year"] >= 2014), ["Population", "Year"]]

#Oregon data
OR_pre_weed = pop_df.loc[(pop_df["State"] == "Oregon") & (pop_df["Year"] < 2016), ["Population", "Year"]]
OR_post_weed = pop_df.loc[(pop_df["State"] == "Oregon") & (pop_df["Year"] >= 2016), ["Population", "Year"]]

#Texas data
TX_pre_weed = pop_df.loc[(pop_df["State"] == "Texas") & (pop_df["Year"] < 2014), ["Population", "Year"]]
TX_post_weed = pop_df.loc[(pop_df["State"] == "Texas") & (pop_df["Year"] >= 2014), ["Population", "Year"]]
```

### Plotting regressions for Colorado

```python
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(CO_pre_weed["Year"], CO_pre_weed["Population"])
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(CO_post_weed["Year"], CO_post_weed["Population"])
line1 = sns.regplot("Year", "Population", CO_pre_weed, line_kws={'label':"before weed" + " "+"y={0:.1f}x+{1:.1f}".format(slope1,intercept1)})
line1.legend()
line1.figure.set_size_inches(12, 8)
sns.despine()
line2 = sns.regplot("Year", "Population", CO_post_weed, line_kws={'label': "after weed" + " "+"y={0:.1f}x+{1:.1f}".format(slope2,intercept2)})
line2.legend()

a = CO_pre_weed["Population"].values.tolist()
b = CO_post_weed["Population"].values.tolist()

# runnning a t-test
tCO, pCO = stats.ttest_ind(a,b)

line1.set_title(f'Colorado population change \n t-statistic = {tCO}, p-value = {pCO}')
plt.savefig("Colorado_pop_change.png")

```

### T-test on California property crime data

```python
t_stats = []
p_values = []
a = crime.loc[(crime["Year"] < 2014), ["Alaska"]]
b = crime.loc[(crime["Year"] >= 2014), ["Alaska"]]
a1 = a['Alaska'].values.tolist()
b1 = b['Alaska'].values.tolist()

# runnning a t-test
tA, pA = stats.ttest_ind(a1,b1)
print("t-statistics = " + str(tA))
print("p = " + str(2*pA))

t_stats.append(tA)
p_values.append(pA)
```
